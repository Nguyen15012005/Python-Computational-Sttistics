### Bài 01

Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import random
import math

# asố câu đúng của bạn sinh viên trong 10 lần kiểm tra
def do_exam():
    dung = 0
    for _ in range(10):
        if random.randint(1, 4) == 1:
            dung += 1
    return dung

print("10 lần làm bài:")
for i in range(10):
    print(f"Lần {i+1}: đúng {do_exam()} câu")

# b xác suất để sinh viên này được 4 điểm
def C(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

n = 10
p = 1/4
k = 4

P = C(n, k) * (p**k) * ((1-p)**(n-k))

print("\nXác suất được 4 điểm:")
print(P)

# c Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)
EX = n * p
VarX = n * p * (1 - p)

print("\nE(X) =", EX)
print("Var(X) =", VarX)
```

### Bài 02

Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

```python
# YOUR CODE HERE
# a. Xác suất để số chấm xuất hiện là 4
N = 5000

count_4 = 0
count_ge_4 = 0 

for _ in range(N):
    roll = random.randint(1, 6) 
    
    if roll == 4:
        count_4 += 1
        
    if roll >= 4:
        count_ge_4 += 1

P_4 = count_4 / N
P_ge_4 = count_ge_4 / N

print("Kết quả mô phỏng:\n")
print(f"Số lần ra 4: {count_4} lần")
print(f"Xác suất để số chấm xuất hiện là 4: {P_4}")

# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
print(f"Số lần >= 4: {count_ge_4} lần")
print(f"Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4: {P_ge_4}")
```

### Bài 03

Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
# YOUR CODE HERE
# a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3
import numpy as np
import matplotlib.pyplot as plt
n = 50
p = 0.7
N = 5000
data = np.random.binomial(n, p, N)
min_val = np.min(data)
max_val = np.max(data)

Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # median
Q3 = np.percentile(data, 75)

print("=== a. Thống kê ===")
print(f"Min = {min_val}")
print(f"Max = {max_val}")
print(f"Q1 = {Q1}")
print(f"Q2 (Median) = {Q2}")
print(f"Q3 = {Q3}")

# b. Vẽ đồ thị boxplot
plt.figure()
plt.boxplot(data)
plt.title("Boxplot - Phân phối nhị thức (n=50, p=0.7)")
plt.ylabel("Giá trị")
plt.show()
# c. Vẽ các đồ thị: 
# tần số
plt.figure()
plt.hist(data)
plt.title("Biểu đồ tần số")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.show()
# tần suất
plt.figure()
plt.hist(data, density=True)
plt.title("Biểu đồ tần suất")
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")
plt.show()
```

### Bài 04

Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. 
Tính xác suất để trung tâm nhận được: 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
# YOUR CODE HERE
lam = 3  
def poisson(k, lam):
    return (lam**k * math.exp(-lam)) / math.factorial(k)

print("Xác suất để trung tâm nhận được:")

for k in [1, 2, 3]:
    P = poisson(k, lam)
    print(f"{k} cuộc gọi: P(X = {k}) = {P:.6f}")
```

### Bài 05

Viết chương trình phát sinh ngẫu nhiên mẫu gồm 10.000 phần tử tuân theo phân phối Poisson với tham số λ=4

a. Vẽ biểu đồ histogram để quan sát phân phối mẫu
b. Tính trung bình và phương sai từ dữ liệu mô phỏng, rồi so sánh với giá trị lý thuyết
c. Thực hiện lại câu a với các giá trị λ lần lượt là: 6, 8, 10, 14, 20. Vẽ trên cùng một hình tất cả các đồ thị histogram.

```python
# YOUR CODE HERE
# a. Vẽ biểu đồ histogram để quan sát phân phối mẫu 
N = 10000
lam = 4

data = np.random.poisson(lam, N)

plt.figure()
plt.hist(data)
plt.title("Histogram - Poisson (λ = 4)")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.show()
# b. Tính trung bình và phương sai từ dữ liệu mô phỏng, rồi so sánh với giá trị lý thuyết 
mean_sim = np.mean(data)
var_sim = np.var(data)

print("So sánh:")
print(f"Trung bình mô phỏng = {mean_sim}")
print(f"Phương sai mô phỏng = {var_sim}")

print(f"Trung bình lý thuyết = {lam}")
print(f"Phương sai lý thuyết = {lam}")

# c. Thực hiện lại câu a với các giá trị λ lần lượt là: 6, 8, 10, 14, 20. Vẽ trên cùng một hình tất cả các đồ thị histogram.
lams = [6, 8, 10, 14, 20]

plt.figure()

for lam in lams:
    data = np.random.poisson(lam, N)
    plt.hist(data, alpha=0.5, label=f"λ = {lam}")

plt.title("Histogram các phân phối Poisson")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.legend()
plt.show()
```
