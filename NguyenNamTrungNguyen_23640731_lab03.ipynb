### Bài 1: Kiểm soát Chất lượng Sản xuất

TÌNH HUỐNG: Một nhà máy có hai dây chuyền sản xuất:

- Dây chuyền A: Chiếm 70% tổng sản lượng, tỷ lệ lỗi 3%

- Dây chuyền B: Chiếm 30% tổng sản lượng, tỷ lệ lỗi 7%

YÊU CẦU:

a) Tính tỷ lệ lỗi chung?

b) Xác suất một sản phẩm là đạt?

c) Dây chuyền sản xuất và chất lượng có độc lập không?

d) Nếu chọn ngẫu nhiên 2 sản phẩm, P(cả hai đều đạt) là bao nhiêu?

```python
# YOUR CODE HERE
# a) Tính tỷ lệ lỗi chung?
p_A = 0.7
p_LoiA = 0.03
p_B = 0.3
p_LoiB = 0.07

p_LoiChung = p_A*p_LoiA + p_B*p_LoiB
print("a) Tỷ lệ lỗi chung là:",round(p_LoiChung, 4), "(", round(p_LoiChung*100,2), "%)")

# b) Xác suất một sản phẩm là đạt?
p_Dat = 1 - p_LoiChung

print("b) Xác suất một sản phẩm đạt là:", p_Dat, "(",p_Dat*100 ,"%)")

# c) Dây chuyền sản xuất và chất lượng có độc lập không?

checkDocLap = p_LoiA == p_LoiChung
if checkDocLap:
    print("c) Dây chuyền sản xuất và chất lượng độc lập")
else:
    print("c) Dây chuyền sản xuất và chất lượng không độc lập")

# d) Nếu chọn ngẫu nhiên 2 sản phẩm, P(cả hai đều đạt) là bao nhiêu?
p_HaiDat = p_Dat**2
print("d) P(cả hai đều đạt) là:", round(p_HaiDat, 4), "(", round(p_HaiDat*100,2), "%)")
```

### Bài 2: Chọn Nhóm Dự án

TÌNH HUỐNG: Một công ty cần chọn nhóm dự án 4 người từ: 8 Kỹ sư, 6 Quản lý, 4 Chuyên viên phân tích.

YÊU CẦU:

a) Có bao nhiêu cách chọn bất kỳ 4 người?

b) Có bao nhiêu cách chọn 2 kỹ sư và 2 quản lý?

c) Xác suất nhóm có ít nhất một chuyên viên phân tích?

d) Xác suất nhóm có đúng một người từ mỗi bộ phận cộng thêm một kỹ sư?

```python
import math
# YOUR CODE HERE
# a) Có bao nhiêu cách chọn bất kỳ 4 người?
def comb(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n-k))

a = comb(18, 4)
print("a) Số cách chọn bất kỳ 4 người là:", a)

# b) Có bao nhiêu cách chọn 2 kỹ sư và 2 quản lý?
b = comb(8, 2) * comb(6, 2)
print("b) Số cách chọn 2 kỹ sư và 2 quản lý là:", b)

# c) Xác suất nhóm có ít nhất một chuyên viên phân tích?
total = comb(18, 4)
khong_A = comb(14, 4)
c = 1 - khong_A / total
print("c) Xác suất có ít nhất 1 chuyên viên phân tích là:", round(c,4), "(",round(c*100,2),"%)")

# d) Xác suất nhóm có đúng một người từ mỗi bộ phận cộng thêm một kỹ sư?
thuan_loi = comb(8, 2) * comb(6, 1) * comb(4, 1)
d = thuan_loi / total
print("d) Xác suất nhóm có đúng một người từ mỗi bộ phận cộng thêm một kỹ sư là:", round(d,4),"(",round(d*100,2),"%)")
```

### Bài 3: Phân tích Hành vi Mua hàng của Khách hàng

TÌNH HUỐNG: Dữ liệu mua hàng của khách hàng:

| Nhóm tuổi | Sản phẩm Cao cấp | Sản phẩm Tiêu chuẩn | Tổng |
| --------- | ---------------- | ------------------- | ---- |
| 18-30     | 0.15             | 0.25                | 0.40 |
| 31-50     | 0.25             | 0.20                | 0.45 |
| 51+       | 0.10             | 0.05                | 0.15 |

YÊU CẦU:

a) P(Mua sản phẩm cao cấp)?

b) P(Nhóm tuổi 31-50)?

c) P(Cao cấp | Tuổi 31-50)?

d) P(Tuổi 31-50 | Cao cấp)?

e) Nhóm tuổi và loại sản phẩm có độc lập không?

```python
# YOUR CODE HERE
# a) P(Mua sản phẩm cao cấp)?
p_cao_cap = 0.15 + 0.25 + 0.10
print("a) P(Mua sản phẩm cao cấp) là:", p_cao_cap)


# b) P(Nhóm tuổi 31-50)?
p_31_50 = 0.25 + 0.20
print("b) P(Nhóm tuổi 31-50) là:", p_31_50)


# c) P(Cao cấp | Tuổi 31-50)?
p_c_given_31_50 = 0.25 / p_31_50
print("c) P(Cao cấp | Tuổi 31-50) là:", round(p_c_given_31_50,2))


# d) P(Tuổi 31-50 | Cao cấp)?
p_31_50_given_c = 0.25 / p_cao_cap
print("d) P(Tuổi 31-50 | Cao cấp) là:", p_31_50_given_c)


# e) Nhóm tuổi và loại sản phẩm có độc lập không?
doc_lap = (p_c_given_31_50 == p_cao_cap)
if doc_lap:   
    print("e) Nhóm tuổi và loại sản phẩm có độc lập là:","(",doc_lap,")")
else:
    print("e) Nhóm tuổi và loại sản phẩm Không độc lập là:","(",doc_lap,")")
```

### Bài 4: Rủi ro Chuỗi Cung ứng

TÌNH HUỐNG: Công ty lấy nguyên liệu từ 3 nhà cung cấp:

A: 50% nguyên liệu, 2% chậm trễ

B: 30% nguyên liệu, 5% chậm trễ

C: 20% nguyên liệu, 8% chậm trễ

YÊU CẦU:
a) Xác suất tổng thể bị chậm trễ?

b) Nếu có chậm trễ, khả năng do nhà cung cấp nào cao nhất?

c) P(giao hàng đúng hạn từ cả 3) nếu đặt hàng từ mỗi nhà cung cấp?



```python
# YOUR CODE HERE
# a) Xác suất tổng thể bị chậm trễ?
p_tre = 0.5 * 0.02 + 0.3 * 0.05 + 0.2 * 0.08
print("a) Xác suất tổng thể bị chậm trễ là:", round(p_tre, 4), "(",round(p_tre*100, 2),"%)")


# b) Nếu có chậm trễ, khả năng do nhà cung cấp nào cao nhất?
# Bayes: P(A|trễ) = P(A)*P(trễ|A) / P(trễ)

p_A_given = 0.5 * 0.02 / p_tre
p_B_given = 0.3 * 0.05 / p_tre
p_C_given = 0.2 * 0.08 / p_tre

print("b)    P(A | trễ) =", round(p_A_given,4), "(",round(p_A_given*100, 2),"%)")
print("      P(B | trễ) =", round(p_B_given,4), "(",round(p_B_given*100, 2),"%)")
print("      P(C | trễ) =", round(p_C_given,4), "(",round(p_C_given*100, 2),"%)")
if p_A_given > p_B_given and p_A_given > p_C_given:
    print("Chậm trễ do nhà cung cấp A là cao nhất", round(p_A_given,4), "(",round(p_A_given*100, 2),"%)")
elif p_B_given > p_C_given:
    print("Chậm trễ do nhà cung cấp B là cao nhất", round(p_B_given,4), "(",round(p_B_given*100, 2),"%)")
else:
    print("Chậm trễ do nhà cung cấp C là cao nhất", round(p_C_given,4), "(",round(p_C_given*100, 2),"%)")

# c) P(giao hàng đúng hạn từ cả 3) nếu đặt hàng từ mỗi nhà cung cấp?
# đúng hạn = 1 - trễ
p_dung_han = (1 - 0.02) * (1 - 0.05) * (1 - 0.08)
print("c) P(giao hàng đúng hạn từ cả 3) nếu đặt hàng từ mỗi nhà cung cấp là: ", round(p_dung_han, 4), "(",round(p_dung_han*100, 2),"%)")
```

### BÀI TẬP 5: Quyết định Ra mắt Sản phẩm

TÌNH HUỐNG: Cây quyết định ra mắt sản phẩm:

Giai đoạn 1 – Nghiên cứu thị trường: 70% tích cực, 30% tiêu cực

Giai đoạn 2a – Nếu tích cực → Ra mắt: 80% thành công, 20% thất bại

Giai đoạn 2b – Nếu tiêu cực → Không ra mắt (0% thành công)

YÊU CẦU:

a) P(sản phẩm thành công)?

b) P(nghiên cứu tích cực | sản phẩm thành công)?

c) Vẽ cây xác suất đầy đủ

```python
# YOUR CODE HERE
# a) P(sản phẩm thành công)?
p_success = 0.7 * 0.8 + 0.3 * 0
print("a) P(sản phẩm thành công) là:", round(p_success, 4))


# b) P(nghiên cứu tích cực | sản phẩm thành công)?
# Bayes: P(TC | success) = P(TC & success) / P(success)
p_tc_given_success = (0.7 * 0.8) / p_success
print("b) P(nghiên cứu tích cực | sản phẩm thành công) là:", round(p_tc_given_success, 4))


# c) Vẽ cây xác suất đầy đủ
print("\nc) Cây xác suất đầy đủ:")
print("Start")
print("├── Tích cực (0.7)")
print("│   ├── Thành công (0.8)  =>", round(0.7 * 0.8, 4))
print("│   └── Thất bại (0.2)    =>", round(0.7 * 0.2, 4))
print("└── Tiêu cực (0.3)")
print("    └── Không ra mắt / Thành công = 0")
```

### BÀI TẬP 6: Xét nghiệm Y tế

TÌNH HUỐNG: Xét nghiệm bệnh hiếm:

- Tỷ lệ mắc: 0.5% dân số
- Độ nhạy: 95%
- Độ đặc hiệu: 98%

YÊU CẦU:

a) Nếu kết quả dương tính, P(có bệnh)?

b) Nếu âm tính, P(khỏe mạnh)?

c) Kết quả thay đổi thế nào nếu tỷ lệ mắc là 10%?

```python
# YOUR CODE HERE
p_benh = 0.005
p_khong_benh = 1 - p_benh

do_nhay = 0.95
do_dac_hieu = 0.98

p_duong_khi_khong = 1 - do_dac_hieu
p_am_khi_benh = 1 - do_nhay


# a) Nếu kết quả dương tính, P(có bệnh)?
p_duong = p_benh * do_nhay + p_khong_benh * p_duong_khi_khong
p_benh_given_duong = p_benh * do_nhay / p_duong
print("a) Nếu dương tính, xác suất có bệnh:",round(p_benh_given_duong, 4),"(",round(p_benh_given_duong * 100, 2),"%)")

# b) Nếu âm tính, P(khỏe mạnh)?
p_am = p_khong_benh * do_dac_hieu + p_benh * p_am_khi_benh
p_khoe_given_am = p_khong_benh * do_dac_hieu / p_am
print("b) Nếu âm tính, xác suất khỏe mạnh:",round(p_khoe_given_am, 4),"(",round(p_khoe_given_am * 100, 2),"%)")

# c) Kết quả thay đổi thế nào nếu tỷ lệ mắc là 10%?
p_benh_new = 0.10
p_khong_benh_new = 1 - p_benh_new

p_duong_new = p_benh_new * do_nhay + p_khong_benh_new * p_duong_khi_khong
p_benh_given_duong_new = p_benh_new * do_nhay / p_duong_new

print("c) Khi tỷ lệ mắc = 10%, nếu dương tính, xác suất có bệnh:",round(p_benh_given_duong_new, 4),"(",round(p_benh_given_duong_new * 100, 2),"%)")
```

### Bài 7: Dự đoán Mua hàng của Khách

TÌNH HUỐNG: Phân khúc khách hàng:
- Cao cấp: 20%, tỷ lệ mua 40%
- Thường: 60%, tỷ lệ mua 15%
- Mới: 20%, tỷ lệ mua 8%

YÊU CẦU:

a) Một khách hàng mua hàng. Họ có khả năng thuộc nhóm nào?

b) Tỷ lệ mua hàng tổng thể?

c) Nên tập trung marketing vào nhóm nào?

```python
# YOUR CODE HERE
# a) Một khách hàng mua hàng. Họ có khả năng thuộc nhóm nào?

p_cao = 0.2
p_thuong = 0.6
p_moi = 0.2

p_mua_cao = 0.4
p_mua_thuong = 0.15
p_mua_moi = 0.08

# xác suất mua tổng
p_mua = p_cao * p_mua_cao + p_thuong * p_mua_thuong + p_moi * p_mua_moi

# Bayes
p_cao_given = p_cao * p_mua_cao / p_mua
p_thuong_given = p_thuong * p_mua_thuong / p_mua
p_moi_given = p_moi * p_mua_moi / p_mua


print(f"a) P(Khách hàng mua Cao Cấp) là: {p_cao_given:.4f} ({p_cao_given*100:.2f}%)")
print(f"   P(Khách hàng mua Thường) là: {p_thuong_given:.4f} ({p_thuong_given*100:.2f}%)")
print(f"   P(Khách hàng mua Mới) là: {p_moi_given:.4f} ({p_moi_given*100:.2f}%)")


# b) Tỷ lệ mua hàng tổng thể?
print(f"b) Tỷ lệ mua hàng tổng thể là: {p_mua:.4f} ({p_mua*100:.2f}%)")


# c) Nên tập trung marketing vào nhóm nào?
if p_cao_given > p_thuong_given and p_cao_given > p_moi_given:
    print("c) Nên tập trung nhóm CAO CẤP")
elif p_thuong_given > p_moi_given:
    print("c) Nên tập trung nhóm THƯỜNG")
else:
    print("c) Nên tập trung nhóm MỚI")
```

### Bài 8: Mô phỏng Độ tin cậy Kỹ thuật

TÌNH HUỐNG: Hệ thống điện tử với 5 linh kiện nối tiếp: Mỗi linh kiện có độ tin cậy 95%. Hệ thống hỏng nếu BẤT KỲ linh kiện nào hỏng

YÊU CẦU:

a) Dùng Monte Carlo ước tính độ tin cậy hệ thống

b) So sánh với công thức lý thuyết

c) Cần bao nhiêu mô phỏng để kết quả chính xác?


```python
# YOUR CODE HERE
import random
# a) Dùng Monte Carlo ước tính độ tin cậy hệ thống
n_sim = 100000   
success = 0

for _ in range(n_sim):
    ok = True
    for _ in range(5):
        if random.random() > 0.95:   
            ok = False
            break
    if ok:
        success += 1

p_monte = success / n_sim

print(f"a) Độ tin cậy ước tính = {p_monte:.4f} ({p_monte*100:.2f}%)")


# b) So sánh với công thức lý thuyết
p_theory = 0.95 ** 5

print(f"b) Độ tin cậy lý thuyết = {p_theory:.4f} ({p_theory*100:.2f}%)")


# c) Sai số giữa mô phỏng và lý thuyết
error = abs(p_monte - p_theory)

print(f"c) Sai số = {error:.6f}")
```

### Bài 9: Rủi ro Danh mục Đầu tư

TÌNH HUỐNG: Danh mục gồm 3 tài sản:

- Cổ phiếu A: 40%, lợi suất kỳ vọng 8%, độ biến động 15%
- Cổ phiếu B: 35%, lợi suất kỳ vọng 12%, độ biến động 20%
- Trái phiếu: 25%, lợi suất kỳ vọng 4%, độ biến động 5%

YÊU CẦU:

a) Mô phỏng lợi suất 1 năm

b) Ước tính Giá trị rủi ro (5th percentile)

c) P(thua lỗ)?

```python
# YOUR CODE HERE
import numpy as np
# a) Mô phỏng lợi suất 1 năm
n_sim = 100000

w_A, w_B, w_T = 0.4, 0.35, 0.25

mean_A, std_A = 0.08, 0.15
mean_B, std_B = 0.12, 0.20
mean_T, std_T = 0.04, 0.05

r_A = np.random.normal(mean_A, std_A, n_sim)
r_B = np.random.normal(mean_B, std_B, n_sim)
r_T = np.random.normal(mean_T, std_T, n_sim)

portfolio_return = w_A*r_A + w_B*r_B + w_T*r_T
print(f"a) Mô phỏng lợi suất 1 năm ≈ {np.mean(portfolio_return):.4f} ({np.mean(portfolio_return)*100:.2f}%)")

# b) Ước tính Giá trị rủi ro (5th percentile)
var_5 = np.percentile(portfolio_return, 5)
print(f"b) Ước tính Giá trị rủi ro (5th percentile) = {var_5:.4f} ({var_5*100:.2f}%)")

# c) P(thua lỗ)?
p_loss = np.mean(portfolio_return < 0)
print(f"c) P(thua lỗ) = {p_loss:.4f} ({p_loss*100:.2f}%)")
```

### Bài 10: Phân tích Độ tin cậy Mạng

TÌNH HUỐNG: Mạng máy tính có đường dự phòng:
- Đường chính: 95% tin cậy
- Đường dự phòng: 90% tin cậy
- Hoạt động độc lập

Hệ thống chạy nếu ít nhất một đường hoạt động

YÊU CẦU:

a) Mô phỏng thời gian hoạt động mạng

b) So sánh đường đơn vs có dự phòng

c) Mức cải thiện tin cậy nhờ dự phòng?


```python
# YOUR CODE HERE
import random
# a) Mô phỏng thời gian hoạt động mạng
n_sim = 100000
success_backup = 0
success_single = 0

for _ in range(n_sim):
    main_ok = random.random() < 0.95
    backup_ok = random.random() < 0.90

    # hệ thống có dự phòng
    if main_ok or backup_ok:
        success_backup += 1

    # chỉ dùng đường chính
    if main_ok:
        success_single += 1


p_backup = success_backup / n_sim
p_single = success_single / n_sim

print(f"a) Kết quả Monte Carlo dự phòng là: {p_backup:.4f} ({p_backup*100:.2f}%)")
print(f"   Kết quả Monte Carlo đường chính là: {p_single:.4f} ({p_single*100:.2f}%)")

# b) So sánh đường đơn vs có dự phòng
print(f"b) Chênh lệch đường chính và có dự phòng là: {p_backup - p_single:.4f} ({(p_backup - p_single)*100:.2f}%)")

# c) Mức cải thiện tin cậy nhờ dự phòng?
improve = (p_backup - p_single) / p_single * 100

print(f"c) Mức cải thiện tin cậy nhờ dự phòng: Tăng {improve:.2f}% so với chỉ dùng 1 đường")
```
